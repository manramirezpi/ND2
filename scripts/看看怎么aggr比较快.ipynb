```python
import numpy as np

N = 1000
E = 2000
V = 200
x = np.random.rand(N, E)
index = np.random.randint(V, size=E)
y = np.zeros((N, V))
```

```python
def scatter_add(x:np.ndarray, index:np.ndarray, y:np.ndarray) -> np.ndarray:
    for k, i in enumerate(index):
        y[..., i] += x[..., k]
    return y
y *= 0
scatter_add(x, index, y)
```

```python
from numba import njit, prange

@njit(parallel=True)
def scatter_add_numba(x: np.ndarray, index: np.ndarray, y: np.ndarray) -> np.ndarray:
    # 假设 x, y 都是 2D，沿第 0 维索引累加
    N, E = x.shape
    for e in prange(E):
        v = index[e]
        for d in range(N):
            y[d, v] += x[d, e]
    return y
y *= 0
scatter_add_numba(x, index, y)
```
